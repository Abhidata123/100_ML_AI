```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Let's see the available datasets in seaborn
sns.get_dataset_names()
```

```python
# Let's use the restaurant tips data available in seaborn
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Univariate Plots

## KDE Plot

```python
# KDE Plot
plt.figure(figsize=(8, 5))
sns.kdeplot(data=tips['total_bill'])
plt.title("Kernel Density Estimation of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Density")
plt.show()
```

### Rug Plot

```python
# Rug Plot
plt.figure(figsize=(8, 5))
sns.rugplot(a=tips['total_bill'], height=0.1)
plt.title("Rug Plot of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Density")
plt.show()
```

## Box Plot

```python
# Box Plot for Univariate Visualization
plt.figure(figsize=(8, 5))
sns.boxplot(x=tips['total_bill'], palette="coolwarm")
plt.title("Total Bill Distribution")
plt.xlabel("Total Bill ($)")
plt.show()
```

## Violin Plot

```python
# Violin Plot for Univariate Visualization
plt.figure(figsize=(8, 5))
sns.violinplot(x=tips['total_bill'], palette="Set2")
plt.title("Total Bill Distribution")
plt.xlabel("Total Bill ($)")
plt.show()
```

## Strip Plot

```python
# Strip Plot for Univariate Visualization
plt.figure(figsize=(8, 5))
sns.stripplot(x=tips['total_bill'], jitter=True)
plt.title("Total Bill Distribution")
plt.xlabel("Total Bill ($)")
plt.show()
```

## Dist Plot

```python
# Strip Plot for Univariate Visualization
plt.figure(figsize=(8, 5))
sns.distplot(a=tips['total_bill'])
plt.title("Total Bill Distribution")
plt.xlabel("Total Bill ($)")
plt.show()
```

## Bi-Variate Plots

### Regression Plot

```python
# Regression Plot
plt.figure(figsize=(8, 5))
sns.regplot(x="total_bill", y="tip", data=tips, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title("Regression Plot of Total Bill vs. Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

### Joint Plot

```python
# Joint Plot
sns.set(style="white")
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")
plt.show()
```

### Hexbin Plot

```python
# Hexbin Plot
plt.figure(figsize=(8, 5))
sns.jointplot(x="total_bill", y="tip",kind='hex', data=tips, gridsize=15, cmap="Blues")
plt.title("Hexbin Plot of Total Bill vs. Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```
